Q: Flatten a nested array.

```python
nested_array = [1,2,3,[4,5,6,[8,9,0]]]
resp = []
def flatten(item):
    for i in item:
        if isinstance(i,list):
            flatten(i)
        else:
            resp.append(i)
    return resp
flatten(nested_array)
print(resp)

#  Time Complexity: O(n)
#  Space Complexity: O(n)
```

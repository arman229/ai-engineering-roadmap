Q: Check if a string is a palindrome.


```python
s = "WeeW"
print(s==s[::-1])

# Time Complexity: O(n)
# Space Complexity: O(n)
```

```python
s = "WeeW"
rev = ''
for i in s:
    rev = i + rev
if rev ==s:
    print('Given String is palindrome')
else:
    print('Give string is not Palindrome')

# Time Complexity: O(n^2)
# Space Complexity: O(n)
```

```python
# This is the most optimized solution to check if a string is palindrome or not.
s = "WeesW"
left, right = 0, len(s)-1
while left < right:
    if s[left]==s[right]:
        left+=1
        right-=1
    else:
        print('Given string is not palindrome')
        break
else:
    print("Given String is palindrome")

# Time Complexity: O(n)
# Space Complexity: O(1)
```

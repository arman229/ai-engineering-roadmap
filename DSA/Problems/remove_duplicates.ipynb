How to Remove Duplicates from an Array?

```python
arr = [3, 4, 2, 1, 2, 2, 1, 4, 5, 6]

seen = set()
result = []

for num in arr:
    if num not in seen:
        seen.add(num)
        result.append(num)

print(result)
# Time Complexity: O(n)
# Space Complexity: O(n)
```

```python
arr = [ 3,4,2,1,2,2,1,4,5,6,7,6,5,4,5,6,7,8]
new_array = []
for i in arr:
    if i not in new_array:
        new_array.append(i)
print(new_array)
# Time Complexity: O(n^2)
# Space Complexity: O(n)
```

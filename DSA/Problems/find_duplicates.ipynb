Q: Find duplicates in an array.

```python
# This is best solution to find duplicates in an array. It uses a set to keep track of seen elements and another set to store duplicates. The time complexity is O(n) and space complexity is O(n) as well.
arr = [8,2,3,3,5,2,3,1,2,3,1,4,5,6,7,8]

seen = set()
duplicates = set()

for num in arr:
    if num in seen:
        duplicates.add(num)
    else:
        seen.add(num)

print(list(duplicates))

# time complexity: O(n)
# space complexity: O(n)

```

```python
arr = [8,2,3,3,5,2,3,1,2,3,1,4,5,6,7,8]

freq = {}

for num in arr:
    freq[num] = freq.get(num, 0) + 1

duplicates = [k for k, v in freq.items() if v > 1]

print(duplicates)

# time complexity: O(n)
# space complexity: O(n)
```

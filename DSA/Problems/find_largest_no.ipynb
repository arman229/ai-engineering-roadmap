Q: Find the largest number in an array.

```python
from math import inf
nu_array = [3,4,3,2,20,44,555]
largest_val = -inf
for i in nu_array:
    if i > largest_val:
        largest_val = i
print(largest_val)
# Time Complexity: O(n)
# Space Complexity: O(1)
```

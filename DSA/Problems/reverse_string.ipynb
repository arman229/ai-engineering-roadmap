Q: Reverse a string.

```python
S = 'My name is ARMAN ASHRAF'
rev=""
for i in S:
    rev = i + rev
print(rev)
    
    

# Time Complexity:  o(n^2)
# Space Complexity:O(n)
```

```python
S = 'My name is ARMAN ASHRAF'
rev = S[::-1]
print(rev)


# Time Complexity: O(n) 
# Space Complexity:O(n)
```

```python
# This function reverses a string in place using two pointers.(best approach)
def reverse_string(s):
    left = 0
    right = len(s) - 1
    s = list(s)

    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1

    return "".join(s)

S = "My name is ARMAN ASHRAF"
print(reverse_string(S))

# Time Complexity: O(n) 
# Space Complexity:O(n)
```

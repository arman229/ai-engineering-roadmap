````python
import pandas as pd
import numpy as np

# Create sample dataset
data = {
    "Name": ["Ali", "Ahmed", "Sara", "Ali", "John", ],
    "Age": [20, 22, 20, 120, 25],  
    "Salary": [50000, 60000, 55000, 50000, 1000000,], 
    "City": ["Lahore", "Karachi", "Lahore", "Lahore", "Islamabad"]
}

df = pd.DataFrame(data)
print(df)
````


# 🧠  Data Transformation

## 📌 Definition:

Data Transformation means:

> **Changing data into a better format so the model can understand it easily**

---

## 🤔 Simple Example

Original data:

```
Salary = 50000
```

Transform:

```
Salary = 50 (in thousands)
```

👉 Same meaning, but easier to work with

---

````python
df['Salary'] = df['Salary'] / 1000  # convert to thousands
print(df)
````


# 🧠  2 Normalization

## 📌 Definition:

Normalization means:

> **Convert values into a range between 0 and 1**

---

## 🤔 Example

Original Age:

```
20, 22, 20
```

After normalization:

```
0.00, 0.02, 0.00
```

---

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df[['Age']])
````

````python
print(df)
````



# 🧠 3. Standardization

## 📌 Definition:

Standardization means:

> **Convert data so that mean = 0 and standard deviation = 1**

---

## 🤔 Example

Original:

```
50000, 60000
```

After standardization:

```
-0.509883, -0.483464
```

---


````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df['Salary'] = scaler.fit_transform(df[['Salary']])
print(df)
````


# 🧠 4. Encoding Categorical Variables

## 📌 Definition:

Encoding means:

> **Convert text (categories) into numbers**

👉 Because ML models only understand numbers

---

## 🔹 Example

```
Gender = Male, Female
```

---

## 🔹 Label Encoding
Assign a number to each category

```
Male → 1  
Female → 0
```
Use Label Encoding when: data has order (low, medium, high)


## 🔹 One-Hot Encoding
Create separate column for each category

| Gender_Male  | Gender_Female |
| -------      | ------        |
| 1            | 0             |
| 0            | 1             | 


````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['City'] = le.fit_transform(df['City'])
print(df)
````

````python
df = pd.get_dummies(df, columns=['Name'])
print(df)
````



# 🧠 5. Feature Scaling

## 📌 Definition:

Feature scaling means:

> **Making all features have similar scale**

---

# 🧠 Example with 3 Data Points

## 📊 Original Data (Before Scaling)

| Person | Age | Salary |
| ------ | --- | ------ |
| A      | 20  | 20000  |
| B      | 25  | 50000  |
| C      | 30  | 80000  |

---

# 🔹 Step 1: Find Min & Max

### Age:

* min = 20
* max = 30

### Salary:

* min = 20000
* max = 80000

---

# 🔹 Step 2: Apply Scaling Formula

$$
x' = \frac{x - min}{max - min}
$$



---

# 🔹 Step 3: Calculate Values

## ✅ Person A

* Age:
  [
  (20 - 20) / (30 - 20) = 0
  ]

* Salary:
  [
  (20000 - 20000) / (80000 - 20000) = 0
  ]

👉 Result: **(0.0, 0.0)**

---

## ✅ Person B

* Age:
  [
  (25 - 20) / (30 - 20) = 5 / 10 = 0.5
  ]

* Salary:
  [
  (50000 - 20000) / (80000 - 20000) = 30000 / 60000 = 0.5
  ]

👉 Result: **(0.5, 0.5)**

---

## ✅ Person C

* Age:
  [
  (30 - 20) / (30 - 20) = 1
  ]

* Salary:
  [
  (80000 - 20000) / (80000 - 20000) = 1
  ]

👉 Result: **(1.0, 1.0)**

---

# 📊 Final Scaled Data

| Person | Age | Salary |
| ------ | --- | ------ |
| A      | 0.0 | 0.0    |
| B      | 0.5 | 0.5    |
| C      | 1.0 | 1.0    |

---
👉 “Scaling makes different features comparable by bringing them into the same range”

# 🧠 7. Correlation Analysis

## 📌 Definition:

**Correlation analysis means:**

> Measuring how strongly two features are related to each other.

---

# 🔹 Simple Meaning

👉 It tells us:

* If one feature increases, does the other increase?
* Or decrease?
* Or no relation?

---

# 📊 Correlation values range from:
$$
-1 \leq r \leq 1
$$
---

## 🔹 What values mean:

| Value | Meaning                  |
| ----- | ------------------------ |
| +1    | Strong positive relation |
| 0     | No relation              |
| -1    | Strong negative relation |

---

# 🤔 Simple Example

## 📈 Positive correlation (+)

```text id="c1"
More study → Higher marks
```

👉 Both increase together

---

## 📉 Negative correlation (-)

```text id="c2"
More speed → Less travel time
```

👉 One increases, other decreases

---

## ⚪ No correlation (0)

```text id="c3"
Age ↔ Shoe color
```

👉 No relation at all


---

# 📊 Example Output

|        | Age | Salary |
| ------ | --- | ------ |
| Age    | 1.0 | 0.8    |
| Salary | 0.8 | 1.0    |

👉 Age and Salary are strongly related

---

**"Why do we need normalization and standardization?"**

We need **normalization** and **standardization** because features in a dataset often have very different scales. If we use them directly, some features can dominate others and negatively affect the performance of many machine learning algorithms.

Consider the following dataset:

| House | Size (sq ft) | Number of Rooms |
| ----- | -----------: | --------------: |
| A     |         1000 |               2 |
| B     |         2000 |               3 |
| C     |         3000 |               4 |

Notice that:

* **Size** ranges from 1000–3000.
* **Number of rooms** ranges from 2–4.

Suppose we want to compare House A and House B using Euclidean distance:

Without scaling:

$$
\sqrt{(2000-1000)^2 + (3-2)^2}
=

\sqrt{1000^2 + 1^2}
\approx 1000
$$

The difference in the number of rooms (1) is almost ignored because the size values are much larger. The algorithm pays attention mainly to the size feature.

After normalization or standardization, both features have similar scales, so each contributes fairly to the model.

### Why do we need them?

#### 1. To give all features equal importance

Without scaling, features with larger numerical values dominate the learning process.

#### 2. To improve the performance of distance-based algorithms

Algorithms such as:

* K-Nearest Neighbors (KNN)
* K-Means Clustering
* Support Vector Machines (SVM)

rely on distances, so they work better when features are on comparable scales.

#### 3. To speed up gradient descent

Algorithms like:

* Linear Regression
* Logistic Regression
* Neural Networks

converge faster and more reliably when the input features are scaled.

#### 4. To prevent numerical instability

Large differences in feature values can make optimization difficult and may lead to poor results.

---

## Normalization

Normalization scales data to a fixed range, usually **0 to 1**:

$$
x' = \frac{x-x_{min}}{x_{max}-x_{min}}
$$

**Use normalization when:**

* The data does not follow a normal distribution.
* You need values in a specific range.
* Working with KNN, K-Means, or Neural Networks.

Example:

```text
Original:     [10, 20, 30, 40, 50]
Normalized:   [0, 0.25, 0.5, 0.75, 1]
```

---

## Standardization

Standardization transforms the data so that it has:

* Mean = 0
* Standard deviation = 1 

**Use standardization when:**

* The data is approximately normally distributed.
* Working with Linear Regression, Logistic Regression, SVM, or PCA.

Example:

```text
Original:       [10, 20, 30, 40, 50]
Standardized:   [-1.41, -0.71, 0, 0.71, 1.41]
```

---
### Difference from normalization

**Normalization** asks:
> "Where does this value lie between the minimum and maximum?"

```text
1000 ---- 2000 ---- 3000
0        0.5        1
```

**Standardization** asks:
> "How far is this value from the average?"

```text
Below Avg     Average     Above Avg
  -1.22          0           +1.22
```

### Summary

| Technique           | Output Range      | Typical Use                                      |
| ------------------- | ----------------- | ------------------------------------------------ |
| **Normalization**   | Usually 0–1       | KNN, K-Means, Neural Networks                    |
| **Standardization** | Mean = 0, Std = 1 | Linear Regression, Logistic Regression, SVM, PCA |

In short, **normalization and standardization are used to ensure that all features contribute fairly and to help machine learning algorithms train more effectively and efficiently.**



# 🧠 5. Feature Scaling

## 📌 Definition:

Feature scaling means:

> **Making all features have similar scale**

---

# 🧠 Example with 3 Data Points

## 📊 Original Data (Before Scaling)

| Person | Age | Salary |
| ------ | --- | ------ |
| A      | 20  | 20000  |
| B      | 25  | 50000  |
| C      | 30  | 80000  |

---

# 🔹 Step 1: Find Min & Max

### Age:

* min = 20
* max = 30

### Salary:

* min = 20000
* max = 80000

---

# 🔹 Step 2: Apply Scaling Formula

$$
x' = \frac{x - min}{max - min}
$$



---

# 🔹 Step 3: Calculate Values

## ✅ Person A

* Age:
  [
  (20 - 20) / (30 - 20) = 0
  ]

* Salary:
  [
  (20000 - 20000) / (80000 - 20000) = 0
  ]

👉 Result: **(0.0, 0.0)**

---

## ✅ Person B

* Age:
  [
  (25 - 20) / (30 - 20) = 5 / 10 = 0.5
  ]

* Salary:
  [
  (50000 - 20000) / (80000 - 20000) = 30000 / 60000 = 0.5
  ]

👉 Result: **(0.5, 0.5)**

---

## ✅ Person C

* Age:
  [
  (30 - 20) / (30 - 20) = 1
  ]

* Salary:
  [
  (80000 - 20000) / (80000 - 20000) = 1
  ]

👉 Result: **(1.0, 1.0)**

---

# 📊 Final Scaled Data

| Person | Age | Salary |
| ------ | --- | ------ |
| A      | 0.0 | 0.0    |
| B      | 0.5 | 0.5    |
| C      | 1.0 | 1.0    |

---
👉 “Scaling makes different features comparable by bringing them into the same range”

# 🧠 7. Correlation Analysis

## 📌 Definition:

**Correlation analysis means:**

> Measuring how strongly two features are related to each other.

---

# 🔹 Simple Meaning

👉 It tells us:

* If one feature increases, does the other increase?
* Or decrease?
* Or no relation?

---

# 📊 Correlation values range from:
$$
-1 \leq r \leq 1
$$
---

## 🔹 What values mean:

| Value | Meaning                  |
| ----- | ------------------------ |
| +1    | Strong positive relation |
| 0     | No relation              |
| -1    | Strong negative relation |

---

# 🤔 Simple Example

## 📈 Positive correlation (+)

```text id="c1"
More study → Higher marks
```

👉 Both increase together

---

## 📉 Negative correlation (-)

```text id="c2"
More speed → Less travel time
```

👉 One increases, other decreases

---

## ⚪ No correlation (0)

```text id="c3"
Age ↔ Shoe color
```

👉 No relation at all


---

# 📊 Example Output

|        | Age | Salary |
| ------ | --- | ------ |
| Age    | 1.0 | 0.8    |
| Salary | 0.8 | 1.0    |

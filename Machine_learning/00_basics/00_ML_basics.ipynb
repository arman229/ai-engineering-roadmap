 

## 🧠 What is Machine Learning?

**Machine Learning (ML)** is a way to make computers **learn from data and make decisions or predictions without being explicitly programmed.**

---

## 🔹 Simple Explanation

Instead of writing rules like:

👉 “If price > 100 then expensive”

You give the computer **data**, and it **learns the rules by itself**.

---

## 🔹 Real-Life Example

### 📌 House Price Prediction

* Input (features):

  * Size
  * Location
  * Number of rooms

* Output (label):

  * Price

👉 The machine learns patterns like:

> “Bigger house → higher price”

--- 

## 🧠 Types of Machine Learning

There are **3 main types of Machine Learning**:

---

# 🔹 1. Supervised Learning

👉 **Definition:**
The model learns from **labeled data** (input + correct output).

### 📌 Example:

* Input: email
* Output: spam / not spam

👉 The model learns:

> “This type of email = spam”

---
 

# 🔹 2. Unsupervised Learning

👉 **Definition:**
The model learns from **data without labels**.

---

### 📌 Example:

* Input: customer data
* Output: groups (clusters)

👉 The model finds patterns like:

> “These customers behave similarly”

---
 

# 🔹 3. Reinforcement Learning

👉 **Definition:**
The model learns by **trial and error using rewards and penalties**.

---

### 📌 Example:

* Game AI
* Robot learning

👉 The model learns:

> “Good action = reward”
> “Bad action = penalty”

---
 


 

# 🧠 1. Dataset

## 📌 Definition:

A **dataset** is a collection of data (usually in table form) used to train a machine learning model.

---

## 🤔 Simple Example:

Think of a dataset like an **Excel sheet**:

| Size | Rooms | Price  |
| ---- | ----- | ------ |
| 1200 | 3     | 200000 |
| 1500 | 4     | 250000 |

👉 This whole table = **dataset**

---

# 🧠 2. Features (Input)

## 📌 Definition:

**Features** are the input variables (the information you give to the model).

---

## 🤔 Simple Example:

From the table:

| Size | Rooms | Price |
| ---- | ----- | ----- |

👉 Features =

* Size
* Rooms

👉 These help predict something

---

## 🎯 Easy Idea:

👉 Features = **inputs / questions**

---

# 🧠 3. Labels (Output)

## 📌 Definition:

**Labels** are the output (the answer you want the model to predict).

---

## 🤔 Simple Example:

From the same table:

| Size | Rooms | Price |
| ---- | ----- | ----- |

👉 Label =

* Price

---

## 🎯 Easy Idea:

👉 Label = **answer**

---

# 🧠 Step 1: Load Dataset (CSV)

```python
import pandas as pd
read_data = pd.read_csv('../Datasets/titanicdataset.csv')
print(read_data.head())
```

# 🧠 Step 2: Understand Dataset (Feature Understanding)
### 📌 Simple Meaning:

* Columns = **features**
* Example features:

  * Age
  * Fare
  * Sex
  * Pclass

```python

print(read_data.info())     # column types
print(read_data.describe()) # statistics
print(read_data.columns)    # column names
```

# 🧠 Step 3: What are Missing Values?

## 📌 Definition:

Missing values = data that is **empty / null**


```python

print(read_data.isnull().sum())
```


## ✅ Handle Missing Values

```python
read_data['Age'] = read_data['Age'].fillna(read_data['Age'].mean())
read_data['Embarked']=read_data['Embarked'].fillna(read_data['Embarked'].mode()[0])
print(read_data.isnull().sum())
```

# 🧠 Step 4: Remove Duplicates

## 📌 Definition:

Duplicate = same row repeated

```python
read_data = read_data.drop_duplicates()
print(read_data.duplicated().sum())
```

# 🧠 Step 5: What are Outliers?

## 📌 Definition:

Outliers = values that are **very different from others**

👉 Example:

* Most ages: 20–50
* One value: 200 ❌ (outlier)


## 🔍 Detect Outliers (Simple)

```python
import matplotlib.pyplot as plt

read_data['Age'].plot(kind='box')
plt.show()
```


## ❌ Remove Outliers (Simple method)

```python
df = read_data[(read_data['Age'] >= 15) & (read_data['Age'] <= 40)]
df['Age'].plot(kind='box')
plt.show()
```

# Python Function
A function is a block of code that performs a specfic task whenever it is called.In bigger programs, where we have large amount of code, it is advisable to create or use existing functions that make the program flow organized and neat.

There are two types of functions:
1. Built-in functions
2. User-defined functions

## 1 Built-in functions:
These functions are defined and pre-coded in python. Some examples of built-in functions are as follows:
min(),max(),print(), etc

## 2 User-defined functions:
We can create functions to perform specfic tasks as per our needs. Such functions are called user-defined functions.

### Syntax:
```python
def function_name(parameters):
    pass
    # Code and statements
```
- Create a function using the def keyword, followed by a function name, followed by a paranthesis (()) and a colon(:).
- Any parameters and arguments should be placed within the parentheses.
- Any Statements and other code within the function should be indented.

### Calling a function:
We call a function by giving the function name, followed by parameters (if any) in the parentheses 

Example:
```python
def print_name(fname,lname):
    print(f"Hello, {fname}, {lname}")
print_name("Arman","Ashraf")
```
Output:
```
Hello, Arman, Ashraf
```

```python
# Dictionaries:
# Dictionaries are very important Python data structure.A Dictionary can contain multiple key/value pairs
# Note: key can be any immutable value like a number, string or tuple.Not the mutable
first_dictionary ={1:1, 2:2, 3:3, 4:4}
second_dictionary ={'a':1, 'b':2, 'c':3, 'd':4}
third_dictionary ={(1,2,4):1, (2,4,6):2}
# fourth_dictionary ={[1,2,4]:1, [2,4,6]:2}
print(first_dictionary)
print(second_dictionary)
print(third_dictionary)
# print(fourth_dictionary)
```

```python
# Note in Dictionary :No duplicate members(key) allowed
second_dictionary ={'a':1, 'b':2, 'c':3, 'd':4,'d':43}
print(second_dictionary)
for i,j in second_dictionary.items():
    print(f'{i}:{j}')
```

```python
# To copy a Dictionary
# Delete a key/value: we can remove a key/value parit from a dictionary using the del statement

second_dictionary ={'a':1, 'b':2, 'c':3, 'd':4,}
del second_dictionary['d']
copy_second_dictionary = second_dictionary.copy() # Shallow copy
copy_second_dictionary['e']=5
print(id(second_dictionary))
print(id(copy_second_dictionary))
print(second_dictionary)
print(copy_second_dictionary)

```

```python
arman_dict = {"name":"arman","age":24}
arman_keys = arman_dict.keys() # it return all the keys in a dict
print(arman_keys)
arman_values = arman_dict.values() # it return all the values in a dict
print(arman_values)
arman_items = arman_dict.items() # it return all the key value pairs
print(arman_items)
single_value = arman_dict['age']
print(single_value)
new_single_value = arman_dict.get('ages','') # if there is no key matches it will return the default value that we pass in single arguments
print(new_single_value)
```

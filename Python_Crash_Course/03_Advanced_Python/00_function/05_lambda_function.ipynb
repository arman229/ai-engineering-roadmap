 
* **Lambda function**: A lambda function is a **small anonymous function** that can take any number of arguments but contains **only one expression**. It is also called an **anonymous function**.
* **Expression**: Produces a **value** when evaluated. Always returns something.
* **Statement**: Performs an **action**. It may or may not produce a value.
 

```python
lambda_fun = lambda x: x+2
print(lambda_fun(3))
```

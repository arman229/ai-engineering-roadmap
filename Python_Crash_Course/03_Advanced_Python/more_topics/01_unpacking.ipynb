#### What does *val mean?
The * operator is called the unpacking operator in Python.
It takes a list (or tuple) and unpacks its elements as individual arguments.

```python
def listofargument(*x): # *x collects all arguments into a tuple
    print(*x) # This unpacks the tuple and prints values separately

x = [1, 2, 3, 4]
listofargument(*x) # * unpacks values from a list/tuple and passes them as separate arguments
```

```python
def listofargument(*x):
    print(type(x))
    print(x)
     
x=[1,2,3]
listofargument(x) 
   
```

```python
a,*b,c = [1,2,3,4,5,6]
print(a)
print(type(b))
print(b)
print(c)
```

👉 In simple words:

* in function definition → collects values into a tuple
* in function call → unpacks values
* in assignment → collects remaining values into a list
